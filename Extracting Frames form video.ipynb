```python
import cv2
import os

def FrameCapture(path, output_folder="frames"):
    # Create folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    vidobj = cv2.VideoCapture(path)

    if not vidobj.isOpened():
        print(f"Error: Cannot open video {path}")
        return

    count = 0
    while True:
        success, image = vidobj.read()
        if not success:
            break  # Stop if no frame is returned

        # Save frame into the folder
        frame_path = os.path.join(output_folder, f"Frame{count}.jpg")
        cv2.imwrite(frame_path, image)
        count += 1

    vidobj.release()
    print(f"Saved {count} frames in folder '{output_folder}'")

# Example usage
if __name__ == '__main__':
    FrameCapture("V1.mp4", output_folder="video_frames")
```

```python
import shutil

# Zip the folder 'video_frames' into 'video_frames.zip'
shutil.make_archive("video_frames", 'zip', "video_frames")
```
